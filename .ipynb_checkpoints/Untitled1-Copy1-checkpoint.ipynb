## Setup

Prerequisites to launch H2O and Flow
* 64 bit Java 6+

### Flow users
1. Download and unpack h2o zip file from website [link](http://h2o-release.s3.amazonaws.com/h2o/rel-wheeler/4/index.html)
1. Run the following command from terminal 
```
cd ~/Downloads
unzip h2o-3.16.0.4.zip
cd h2o-3.16.0.4
java -jar h2o.jar
```
1. Point your browser to http://localhost:54321

### R users
1. Prerequisite: R installed (version 3 or later)
1. Install from cran
```
install.packages(“statmod”)
install.packages(“RCurl”)
install.packages(“jsonlite”)
install.packages(“h2o”)
```
1. Check that library is properly installed:
```
library(h2o)
h2o.init(nthreads = -1)
```

## Small demo with MNIST dataset
![MNIST](http://i.ytimg.com/vi/0QI3xgXuB-Q/hqdefault.jpg)

````python
help(h2o)
````

````python
from h2o.estimators.deepwater import H2ODeepWaterEstimator
````

````python
test_df = h2o.import_file("https://s3.amazonaws.com/h2o-public-test-data/bigdata/laptop/mnist/test.csv.gz")
train_df = h2o.import_file("https://s3.amazonaws.com/h2o-public-test-data/bigdata/laptop/mnist/train.csv.gz")
````

````python
test_df.describe()
````

````python
# specify response and predictor columns 
y = "C785"
x = train_df.names[0:784]
````

````python
# ?train_df.asfactor() -> Convert columns in the current frame to categoricals.
train_df[y] = train_df[y].asfactor()
test_df[y] = test_df[y].asfactor()
````

````python
from h2o.estimators.deepwater import H2ODeepWaterEstimator
help(H2ODeepWaterEstimator)
````

````python
model = H2ODeepWaterEstimator(
    model_id             = "deeplearning-PYTHON",
    distribution         = "multinomial",
    activation           = "rectifier",
    mini_batch_size      = 128,
    hidden               = [1024,1024],
    hidden_dropout_ratios= [0.5,0.5],      ## for better generalization
    input_dropout_ratio  = 0.1,
    sparse               = True,           ## can result in speedup for sparse data
    epochs               = 50)             ## need more epochs for a better model
````

````python
# help(H2ODeepWaterEstimator)
model.train(
    x=x, 
    y=y,
    training_frame=train_df,
    validation_frame=test_df
)
````

````python
model.model_performance(train=True) # training metrics
````

````python
model.model_performance(valid=True) # validation metrics
````
